re is Python‚Äôs built-in module for working with Regular Expressions (regex).

In simple terms:
üëâ re helps you search, match, extract, replace, or clean text using patterns.

This is super important in NLP, spam detection, and text preprocessing.

üîπ Common re functions

1Ô∏è‚É£ re.sub() ‚Üí Replace

```python
import re

text = "Hello!!! How are you???"
clean = re.sub(r"[!?]", "", text)
print(clean)
```

2Ô∏è‚É£ re.findall() ‚Üí Extract

```python
text = "Contact me at test@gmail.com or admin@yahoo.com"
emails = re.findall(r"\S+@\S+", text)
print(emails)
```

3Ô∏è‚É£ re.search() ‚Üí Check if pattern exists

```python
sentence="get free rewards"
if re.search(r"free", sentence):
    print("Found")
```

4Ô∏è‚É£ re.split() ‚Üí Split using pattern

```python
sentence="This   is    NLP"
re.split(r"\s+", sentence)
```
